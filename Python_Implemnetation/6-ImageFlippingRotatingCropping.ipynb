```python
# Import for OpenCV
import cv2

import numpy as np
import matplotlib.pyplot as plt 
```

```python

# Load the image and get it as BGR
image = cv2.imread("./Giraf.jpg")
print(image)
```

```python
# convert to RGB image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

Flip the Image


```python
# Choose 1  flip image Horizontally 
# Choose 0  flip image Vertically 
# Choose -1 flip image on both axes  
flip_horizontal = cv2.flip(image, 1)
flip_vertical   = cv2.flip(image, 0) 
flip_both       = cv2.flip(image, -1)
```

```python
# Plot
titles = [ "Original image" , "Horizontal Flip", "Vertical Flip", "Flip Both axes" ]
images = [image_rgb,
                    cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB),
                    cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB),
                    cv2.cvtColor(flip_both, cv2.COLOR_BGR2RGB)]

plt.figure(figsize=(12,12))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
