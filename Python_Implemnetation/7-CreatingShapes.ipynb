```python
# Import for OpenCV
import cv2

import numpy as np
import matplotlib.pyplot as plt 
```

## Draw a square 

```python


# Create empty canvas
height, width = 600, 800
# Line Color 
blue = (255, 120, 0)

# Create the canvas, using numpy (hight, width, nu of channels), LineColor, data type of values
# Here we will get a numpy array with blue values and with dimensions of width * height
canvas = np.full((height, width, 3), blue, dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

## Draw a line

```python
# will draw the line on top of the square

start_coordinates = (50,50)
end_coordinates = (650, 500)
line_color = (255, 20, 255) # red color
line_thickness = 3

cv2.line(canvas, start_coordinates, end_coordinates, line_color, line_thickness)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

## Draw a rectangle

```python

start_coordinates = (120,100)
end_coordinates = (650, 500)
line_color = (0, 255, 255) # yellow color
line_thickness = 5

cv2.rectangle(canvas, start_coordinates, end_coordinates, line_color, line_thickness)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

## Rectangle with full background

```python

start_coordinates = (200,250)
end_coordinates = (500, 350)
line_color = (2, 70, 255) # yellow color
line_thickness = -1

cv2.rectangle(canvas, start_coordinates, end_coordinates, line_color, line_thickness)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

## Draw a Circle 

```python

center_coordinates = (400,250)
radius = 100
line_color = (200, 30, 255) # yellow color
line_thickness = -1 # To fill the circle

cv2.circle(canvas, center_coordinates, radius, line_color, line_thickness)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

## Draw an Ellipse

```python
center_coordinates = (400,250)
major_axis = 100
minor_axis = 50

angle = 43
start_angle = 100
end_angle  = 50

line_color = (40, 30, 255) # yellow color
line_thickness = 5

cv2.circle(canvas, center_coordinates, (major_axis,minor_axis), angle, start_angle, end_angle, line_color, line_thickness)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```
