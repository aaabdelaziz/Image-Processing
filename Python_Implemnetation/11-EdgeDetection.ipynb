```python
# Import for OpenCV
import cv2

import numpy as np
import matplotlib.pyplot as plt 
```

```python

# Load the image and get it as greyscale
image = cv2.imread("./images/giraf.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap='grey')
```

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Laplacian
laplacian = cv2.Laplacian(image, cv2.CV_64F, ksize=3)
laplacian_abs = cv2.convertScaleAbs(laplacian)

# Canny
lower_threshold = 10
upper_threshold = 200
canny = cv2.Canny(image, lower_threshold, upper_threshold)
```

```python
laplacian
laplacian_abs
```

```python
titles = [ "Original image" , "Sobel", "Laplacian_all", "Laplacian_abs", "Canny"]
images = [image, sobel_combined, laplacian, laplacian_abs, canny]

plt.figure(figsize=(12,12))
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i], cmap='grey')
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
