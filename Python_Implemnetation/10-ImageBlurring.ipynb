```python
# Import for OpenCV
import cv2

import numpy as np
import matplotlib.pyplot as plt 
```

```python

# Load the image and get it as greyscale
image = cv2.imread("./images/giraf.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Average blur
kernel_s = 10
average_blur = cv2.blur(image_grey, (kernel_s, kernel_s))

# Apply cmap to convert to grey
plt.imshow(average_blur, cmap='grey')
```

```python
# Gaussian blur
# Note to use lower kernel size, otherwise you will get failed) ksize.width > 0 && ksize.width % 2 == 1 && ksize.height > 0 && ksize.height % 2 == 1
kernel_s = 3 

gaussian_blur = cv2.GaussianBlur(image_grey, (kernel_s, kernel_s), 0)

# Apply cmap to convert to grey
plt.imshow(gaussian_blur, cmap='grey')
```

```python
# Median blur
# Note to use lower kernel matrix size, otherwise you will get failed) ksize.width > 0 && ksize.width % 2 == 1 && ksize.height > 0 && ksize.height % 2 == 1
kernel_s = 3 

median_blur = cv2.medianBlur(image_grey, kernel_s)

# Apply cmap to convert to grey
plt.imshow(median_blur, cmap='grey')
```

```python
# Show all resized images with each algorithm
plt.figure(figsize=(10,10))

# Show the original Image
plt.subplot(2,2,1)
plt.imshow(image_grey, cmap='grey')
plt.title("Original Image")
plt.axis("off")

# Show the Horizontal Edges Image
plt.subplot(2,2,2)
plt.imshow(average_blur, cmap='grey')
plt.title("Average Blur Image")
plt.axis("off")

# Show the Vertical Edges Image
plt.subplot(2,2,3)
plt.imshow(gaussian_blur, cmap='grey')
plt.title("Gaussian Blur Image")
plt.axis("off")

# Show the Diagonal Edges Image
plt.subplot(2,2,4)
plt.imshow(median_blur, cmap='grey')
plt.title("Median Blur Image")
plt.axis("off")
```
