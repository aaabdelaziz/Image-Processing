```python
# Import for OpenCV
import cv2

import numpy as np
import matplotlib.pyplot as plt 
```

```python

# Load the image and get it as BGR
image = cv2.imread("./Giraf.jpg")
```

```python
# convert to RGB image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
# get the image dimensions (height, width, num of channels(num of colors))
print(image_rgb.shape) 

# Note that image_rgb.shape[0] is the height
# Note that image_rgb.shape[1] is the width
```

## Image Resizing

```python
# Resize to new dimensions
new_height, new_width = 500, 500

resized_image = cv2.resize(image_rgb, (new_width, new_height))
plt.imshow(resized_image)
```

## Image Scaling

```python
# Image scaling
scale_percentage = 30 # Scale by 30%

# Scale dimensions
s_width = int(image_rgb.shape[1] * scale_percentage / 100)
s_height = int(image_rgb.shape[0] * scale_percentage / 100)

# Create the new scaled image, choose which interpolation algorithm to use (INTER_AREA, INTER_LINEAR,..)
scaled_image = cv2.resize(image_rgb, (s_width, s_height), interpolation=cv2.INTER_AREA)

# Print the new scaled image dimensions
print(scaled_image.shape)

plt.imshow(scaled_image)
plt.title("scaled Image")

# The images are created with the help of algorithms based on Neworal network
```

## Image Interpolation

```python
resize_width = 600
resize_height = 600

# Interpolation methods Dictionary
methods = {
    'Nearst' : cv2.INTER_NEAREST,
    'Linear' : cv2.INTER_LINEAR,
    'Area' : cv2.INTER_AREA,
    'Cubic' : cv2.INTER_CUBIC,
    'Lanczos4' : cv2.INTER_LANCZOS4
}

resized_images = {}

# Create resized images for every algorithm
for name, method in methods.items():
    resized_images[name] = cv2.resize(image_rgb, (resize_width,resize_height), interpolation=method)
    # Create image for every resized algorithm
    cv2.imwrite(f"./{name}.jpg", resized_images[name])
```

```python
# Show all resized images with each algorithm
plt.figure(figsize=(12,8))

# Show the original Image
plt.subplot(2,3,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

# Show the resized images, start from subplot (2)
for i, (name, resized_image) in enumerate(resized_images.items(), start=2):
    plt.subplot(2,3,i)
    plt.imshow(resized_image)
    plt.title(name)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
