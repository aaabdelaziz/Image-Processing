```python
# Import for OpenCV
import cv2

import numpy as np
import matplotlib.pyplot as plt 
```

```python

# Load the image and get it as BGR
image = cv2.imread("./Giraf.jpg")
```

```python
# convert to RGB image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
# get the image dimensions (height, width, num of channels(num of colors))
print(image_rgb.shape) 

# Note that image_rgb.shape[0] is the height
# Note that image_rgb.shape[1] is the width
```

```python
# Resize to new dimensions
new_height, new_width = 500, 500

resized_image = cv2.resize(image_rgb, (new_width, new_height))
plt.imshow(resized_image)
```

```python
# Image scaling
scale_percentage = 30 # Scale by 30%

# Scale dimensions
s_width = int(image_rgb.shape[1] * scale_percentage / 100)
s_height = int(image_rgb.shape[0] * scale_percentage / 100)

# Create the new scaled image, choose which interpolation algorithm to use (INTER_AREA, INTER_LINEAR,..)
scaled_image = cv2.resize(image_rgb, (s_width, s_height), interpolation=cv2.INTER_AREA)

# Print the new scaled image dimensions
print(scaled_image.shape)

plt.imshow(scaled_image)
plt.title("scaled Image")

# The images are created with the help of algorithms based on Neworal network
```
