```python
# Import for OpenCV
import cv2

import numpy as np
import matplotlib.pyplot as plt 
```

```python

# Load the image and get it as greyscale
image = cv2.imread("./lines.jpg", cv2.IMREAD_GRAYSCALE)
# image = cv2.imread("./images/giraf.jpg", 0) # or just put 0 

# Apply cmap to convert to grey
plt.imshow(image, cmap='grey')
```

```python
# Change the intensity
# as we increase x negativity, the edges becomes more clear.
x= -15

# Define the horizontal kernel to detect the horizontal edges in the image
# Note:  The negative values detect transitions from dark to bright areas horizontally.
horizontal_kernel = np.array([[-1, -1, -1],
                              [x, x, x],
                              [1, 1, 1]])

# Define the vertical kernel to detect the vertical edges in the image (Rotate the horizontal matrix; rows becomes cols)
vertical_kernel = np.array([[-1, x, 1],
                            [-1, x, 1],
                            [-1, x, 1]])

# Define the diagonal kernel to detect the diagonal edges in the image (below diagonal +1s, above -1s)
diagonal_kernel = np.array([[x, x-1, -1],
                            [1, x, -1],
                            [1, 1, x]])

# apply convolution operation applies the convolution operation using the defined kernel to the input image. 
# The -1 argument indicates that the output image has the same depth as the source.
# This will multiply the image pixels by the filter
horizontal_filtered_image = cv2.filter2D(image, -1, horizontal_kernel)
vertical_filtered_image = cv2.filter2D(image, -1, vertical_kernel)
diagonal_filtered_image = cv2.filter2D(image, -1, diagonal_kernel)
```

```python
# Show all resized images with each algorithm
plt.figure(figsize=(10,10))

# Show the original Image
plt.subplot(2,2,1)
plt.imshow(image, cmap='grey')
plt.title("Original Image")
plt.axis("off")

# Show the Horizontal Edges Image
plt.subplot(2,2,2)
plt.imshow(horizontal_filtered_image, cmap='grey')
plt.title("Horizontal Edges Image")
plt.axis("off")

# Show the Vertical Edges Image
plt.subplot(2,2,3)
plt.imshow(vertical_filtered_image, cmap='grey')
plt.title("Vertical Edges Image")
plt.axis("off")

# Show the Diagonal Edges Image
plt.subplot(2,2,4)
plt.imshow(diagonal_filtered_image, cmap='grey')
plt.title("Diagonal Edges Image")
plt.axis("off")
```
